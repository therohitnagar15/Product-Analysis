```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("data.csv")
```

```python
df.head()
```

```python
df.info()
df.shape
df.describe()
```

```python
df.isnull().sum()
```

```python
df['Date']=pd.to_datetime(df["Date"], format = "%d-%m-%Y")
```

```python
df['Date'].dtype
```

```python
df["Total Sales"] = df["Price"] * df["Quantity"]
```

```python
print("Total Revenue: ",df["Total Sales"].sum())
```

```python
category_sales = df.groupby('Category')['Total Sales'].sum()
print(category_sales)
```

```python
category_sales.plot(kind = "bar")
plt.title("Sale By Category")
plt.ylabel("Total Sale ($)")
plt.show()
```

```python
top_products = df.groupby('Product')['Total Sales'].sum().sort_values(ascending = False)
print(top_products)
```

```python
top_products.plot(kind = "bar")
plt.title("Top Product")
plt.ylabel(f"Total Sale ($)")
plt.show()
```

```python
city_sales = df.groupby('City')['Total Sales'].sum()
print(city_sales)
```

```python
city_sales.plot(kind = "bar")
plt.title("Sales by City")
plt.ylabel("Total Sales ($)")
```

```python
daily_sales = df.groupby('Date')['Total Sales'].sum()
print(daily_sales)
```

```python
daily_sales.plot(kind="bar")
plt.title("Daily Sales")
plt.ylabel("Total Sales ($)")
```
